```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np

img = cv.imread('/content/ktm_uts.jpg', cv.IMREAD_COLOR)
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# Sharpening Filter
sharpen_kernel = np.array([[0, -1, 0],
                           [-1,  5, -1],
                           [0, -1, 0]])
sharpened = cv.filter2D(img_rgb, -1, sharpen_kernel)

# Contrast and Brightness
enhanced = cv.convertScaleAbs(sharpened, alpha=3, beta=25)

# Plot Image
plt.figure(figsize=(20, 10))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')

# After Sharpening Image
plt.subplot(1, 3, 2)
plt.imshow(sharpened)
plt.title("Sharpening Filter")
plt.axis('off')

# Enhanced Image (after brightness and contrast)
plt.subplot(1, 3, 3)
plt.imshow(enhanced)
plt.title("Brightness/Contrast")
plt.axis('off')

plt.show()
```
