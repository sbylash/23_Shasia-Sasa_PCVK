```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
cascade_wajah = cv.CascadeClassifier('/content/drive/MyDrive/Images/haarcascades/haarcascade_frontalcatface.xml')
cat = cv.imread('/content/drive/MyDrive/Images/Object Detection/cat2_templatejpg.jpg',0)
roi_wajah = cascade_wajah.detectMultiScale(cat)
for(x,y,w,h) in roi_wajah:
 cv.rectangle(cat,(x,y),(x+w,y+h),(255,255,255),3)
plt.figure(figsize = (10,10))
plt.imshow(cat, cmap='gray')
```
